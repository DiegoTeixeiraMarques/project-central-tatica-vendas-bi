#### 0. Funções

```python
def anonimizar_numero(codigo):
  return abs(int(((((codigo * 0.6) - 40) + 23) - 2) * 2.7))

def anonimizar_descricao(codigo):
  return "Produto " + str(anonimizar_numero(codigo))
```

### 1. Carregar Dados

#### 1.1. dProduto

```python
import pandas as pd
# Produtos
caminho_arquivo = '/content/drive/My Drive/Projetos/[PowerBI] Central de Vendas/dProdutos.xlsm'
dados = pd.read_excel(caminho_arquivo)
dados.tail(3)
```

```python
dados.info()
```

```python
list_idsku = dados['IdSku']
lista_nova_idsku = []
lista_nova_nmidsku = []
for i in list_idsku:
  lista_nova_idsku.append(anonimizar_numero(i))
  lista_nova_nmidsku.append(anonimizar_descricao(i))

print(lista_nova_idsku[0:3])
print(lista_nova_nmidsku[0:3])
```

```python
novos_dados = dados.copy()

novos_dados['IdSku'] = lista_nova_idsku
novos_dados['NmSku'] = lista_nova_nmidsku
novos_dados['NmSkuId'] = novos_dados['IdSku'].map(str) + ' | ' + novos_dados['NmSku'].map(str)
```

```python
caminho_arquivo = '/content/drive/My Drive/Projetos/[PowerBI] Central de Vendas/dProdutos_novo.csv'
novos_dados.to_csv(caminho_arquivo, index=False)
```
